```python
from evolution import epsilon, epsilon_PP, epsilon_CNO
import astropy.units as u
import numpy as np
import matplotlib.pyplot as plt

rho = 1 * u.g / u.cm**3
X = 0.98
T = np.logspace(5, 10) * u.K
e_PP = epsilon_PP(rho, T, X)
e_CNO = epsilon_CNO(rho, T, X)

plt.plot(T, e_PP, ".", label="PP")
plt.plot(T, e_CNO, ".", label="CNO")
plt.plot(T, e_PP + e_CNO, color="black", label="PP+CNO")
plt.xscale("log")
plt.yscale("log")
plt.legend()
plt.ylabel("$\\epsilon[cm2/s3]$")
plt.xlabel("T[K]")
```
