```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
M1 = pd.read_csv("output/M1.0.csv")
M0_01 = pd.read_csv("output/M0.01.csv")
M0_5 = pd.read_csv("output/M0.5.csv")
M30 = pd.read_csv("output/M30.0.csv")
```

```python
plt.plot(M1["T_eff"], M1["L"], ".", label="$M=1[M_\\odot]$")
plt.plot(M0_5["T_eff"], M0_5["L"], ".", label="$M=0.5[M_\\odot]$")
plt.plot(M0_01["T_eff"], M0_01["L"], ".", label="$M=0.01[M_\\odot]$")
plt.plot(M30["T_eff"], M30["L"], ".", label="$M=30[M_\\odot]$")
plt.xlabel("$T_{eff} [K]$")
plt.ylabel("$L [L_\\odot]$")
plt.xscale("log")
plt.yscale("log")
plt.legend()
plt.xlim(100, 10e6)
plt.ylim(10e-8, 10e5)
plt.title("Stellar Evolution")
plt.gca().invert_xaxis()
```

```python
plt.plot(M0_01["t"], M0_01["R"], ".")
plt.plot(M0_5["t"], M0_5["R"], ".")
plt.plot(M1["t"], M1["R"], ".")
plt.plot(M30["t"], M30["R"], ".")
plt.yscale("log")
plt.xscale("log")
plt.xlabel("$t[yr]$")
plt.ylabel("$R[R_\\odot]$")
plt.
```
